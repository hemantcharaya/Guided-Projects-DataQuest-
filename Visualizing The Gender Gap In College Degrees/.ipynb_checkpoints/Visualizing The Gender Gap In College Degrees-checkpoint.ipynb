# Visualizing The Gender Gap in College Degrees
In this guided project we will try to analyze how the status of gender gap in college degrees is taking shape over the years. Our data consists of the percentage of a particular degree awarded to women in a given year. Here is the list of columns we have in the dataset:

* Year                              (int64)
* Agriculture                       (float64)
* Architecture                      (float64)
* Art and Performance               (float64)
* Biology                           (float64)
* Business                          (float64)
* Communications and Journalism     (float64)
* Computer Science                  (float64)
* Education                         (float64)
* Engineering                       (float64)
* English                           (float64)
* Foreign Languages                 (float64)
* Health Professions                (float64)
* Math and Statistics               (float64)
* Physical Sciences                 (float64)
* Psychology                        (float64)
* Public Administration             (float64)
* Social Sciences and History       (float64)


```python
# Let's first import all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# reading the csv file
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

We will start by categorizing degrees into three different categories:
(1) STEM Category (stem_cats) 
(2) Liberal Arts Category (lib_arts_cats)
(3) Other Catgory (other_cats)

Then we will plot curves for percentage of degrees awarded to men and women over the years and try to analyze the gender gap for each category seperately

```python
# Categorizing the degress in different categories

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# Defining the color which we'll use in our plots

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255) 

# We will transport all our subplots in a figure(defined below)

fig = plt.figure(figsize = (9, 18))

# Now we will start plotting

for sp in range(0,17):
    if sp<6: # the first 6 subplots willl come from stem_cats
        ax = fig.add_subplot(6,3,3*(sp)+1) 
        
        # 3(sp)+1 will set the index where that subplot will be placed in the figure
        # This is done so that subplots from each category will fall under same column
        
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
        
        # women_degrees["<col>"] is percentage of women getting that degre; 
        # so 100-women_degrees["<col>"] will be the percentage of men awarded that degreeduring that year
        
        for key, spine in ax.spines.items():
            spine.set_visible(False)
            
        # The above code is just removing the spine from our subplots to make them much easier to read
        
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(stem_cats[sp])
        ax.set_yticks([0,100])
        ax.axhline(50, c =(171/255, 171/255, 171/255), alpha = 0.3)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")    
        if sp == 0:
            ax.text(2011,100 - women_degrees.loc[41,stem_cats[sp]], 'Men')
            ax.text(2011, women_degrees.loc[41,stem_cats[sp]], 'Women')
        if sp == 5:
            ax.text(2011,100 - women_degrees.loc[41,stem_cats[sp]], 'Men')
            ax.text(2011, women_degrees.loc[41,stem_cats[sp]], 'Women')
            ax.tick_params(labelbottom = "on")
            
    if sp >= 6 and sp<12: # the next 6 subplots will come from other_cats
        ax = fig.add_subplot(6,3,3*(sp-6)+2)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[sp-6]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp-6]], c=cb_orange, label='Men', linewidth=3)
        for key, spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(other_cats[sp-6])
        ax.set_yticks([0,100])
        ax.axhline(50, c =(171/255, 171/255, 171/255), alpha = 0.3)
        ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom = "off")    
        if sp == 6:
            ax.text(2011,100 - women_degrees.loc[41,other_cats[sp-6]], 'Men')
            ax.text(2011, women_degrees.loc[41,other_cats[sp-6]], 'Women')
        if sp == 11:
            ax.text(2011,100 - women_degrees.loc[41,other_cats[sp-6]], 'Men')
            ax.text(2011, women_degrees.loc[41,other_cats[sp-6]], 'Women')
            ax.tick_params(labelbottom = "on")
            
    if sp >= 12 and sp <17: # the last 5 subplots will come from lib_arts_cats
        ax = fig.add_subplot(6,3,3*(sp-12)+3)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp-12]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp-12]], c=cb_orange, label='Men', linewidth=3)
        for key, spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(lib_arts_cats[sp-12])
        ax.set_yticks([0,100])
        ax.axhline(50, c =(171/255, 171/255, 171/255), alpha = 0.3)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")
        if sp == 12:
            ax.text(2011,100 - women_degrees.loc[41,lib_arts_cats[sp-12]], 'Men')
            ax.text(2011, women_degrees.loc[41,lib_arts_cats[sp-12]], 'Women')
        if sp == 16:
            ax.text(2011,100 - women_degrees.loc[41,lib_arts_cats[sp-12]], 'Men')
            ax.text(2011, women_degrees.loc[41,lib_arts_cats[sp-12]], 'Women')
            ax.tick_params(labelbottom = "on")

plt.show()
```

```python
plt.savefig("gender_degrees.png")
# Saving our figure
```

# STEM Fields

Still there is huge gender gap exists among STEm major, particularly with "Engineering" and "Computer Science". On the other hand for "Maths and Statistics" & "Physical Sciences" the gender gap is slowly decreasing. "Biology" and "Psychology" are much prefeerred STEM majors among women whihch is why there is gender gap in favor of women in respective fileds. 

# Liberal Arts

Except for "Social Sciences and History" women are in majority among all majors falling under Liberal Arts. 

# Other Categories 

Women are in majority in "Health Profession", "Public Administartion" and "Education". Whereas, over the years more women have opeted for "Agriculture", "Architecture" and "Business" & thus closing the gender gap
