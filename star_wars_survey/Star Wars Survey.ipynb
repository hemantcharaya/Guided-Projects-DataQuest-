# Star Wars Survey 
In this project we'll use the star wars survey to practice data cleaning and make it ready for the data analysis. Later we will try to find out how the preference for different Star Wars movies varies among Male and Female 

```python
# We'll start by importing all the required files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
# start analyzing the data with first 10 rows
star_wars.head(10)
```

- We should remove any NaN values from "RespondentID" column since "ResponsdentID" column showld contain unique and non null values
- Below we will see that after removing null values, the total unique values (in "RespondentID") are 1186 (same as no. of entries). This implies that there are not any repeated values in "RespondentID" column 

```python
print(star_wars.shape)
star_wars = star_wars.dropna(subset = ["RespondentID"])
star_wars["RespondentID"].unique
```

```python
star_wars.head(10)
```

### Converting columns with values "Yes", "No" to Boolean using Mapping & Dictionaries

```python
# creating the dictionary for mapping
yes_no = {
    "Yes":True,
    "No":False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)


star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

- The next 6 columns is basially the response of the respondents to a question about which of 6 star wars movies they have watched
- Our next step would be to rename the columns' name (rename it with somehting intuitive) and convert the values in the columns into a Boolean (True or False) with mapping as we did above  

```python
star_wars.rename({"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 4":"seen_2"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 5":"seen_3"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 6":"seen_4"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 7":"seen_5"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 8":"seen_6"}, axis = 1, inplace = True)
```

```python
# Now we will convert the values of these columns into Boolean
map_seen = {
    "Star Wars: Episode I  The Phantom Menace":True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.NaN:False
}

seen = ["seen_1", "seen_2", "seen_3", "seen_4","seen_5","seen_6"]
for P in seen:
     star_wars[P] = star_wars[P].map(map_seen)

star_wars.head(10)
```

- In the next 6 columns contain the respondant has ranked the 6 star wars movies from 1 to 6 
- Again, there are two eopartin we will be conducting here:
(1) renaming these columns to something more descrptive (ranking_1, ranking_2 ....) and 
(2) We will convert the data type of these columns 

```python
# Renaming the rows
star_wars.rename({"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 10":"ranking_2"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 11":"ranking_3"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 12":"ranking_4"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 13":"ranking_5"}, axis = 1, inplace = True)
star_wars.rename({"Unnamed: 14":"ranking_6"}, axis = 1, inplace = True)
# Changing the data type
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)
```

- We would like to know to what extent each movie was liked by respondents
- To do so, we'll take the mean of all the ranking given by the respondents


```python
rank = star_wars.mean()
ax_1 = plt.subplot()
bar_heights = [rank["ranking_1"], rank["ranking_2"], rank["ranking_3"], rank["ranking_4"], rank["ranking_5"], rank["ranking_6"]]
bar_position =[]
x_tickslabels = ["Star Wars I", "Star Wars II","Star Wars III","Star Wars IV","Star Wars V","Star Wars VI"]
x_tickposition = []
for P in range(0,6):
    Q = P/2+0.1
    bar_position.append(Q)
    
for P in range(0,6):
    Q = P/2+0.1
    x_tickposition.append(Q)

ax_1.bar(bar_position, bar_heights, 0.3)
ax_1.set_xticks(x_tickposition)
ax_1.set_xticklabels(x_tickslabels, rotation = 90)
ax_1.set_xlabel("Movies")
ax_1.set_ylabel("Avg. Rating")
```

- Lesser the mean ranking implies more people liked that movie
- From the bar chart above we can safely order the movies on the basis of respondents' linking:

Star Wars V > Star Wars VI > Star Wars IV > Star Wars I > Star Wars II > Star Wars III

-  Next step will be to count the number of respondents watching each movie 
- To do so, we'll take the sum of all the seen[i] columns in our dataset 

```python
num_watch = star_wars.sum()
ax_2 = plt.subplot()
bar_heights = [num_watch["seen_1"], num_watch["seen_2"], num_watch["seen_3"], num_watch["seen_4"], num_watch["seen_5"], num_watch["seen_6"]]
bar_position =[]
x_tickslabels = ["Star Wars I", "Star Wars II","Star Wars III","Star Wars IV","Star Wars V","Star Wars VI"]
x_tickposition = []
for P in range(0,6):
    Q = P/2+0.1
    bar_position.append(Q)
    
for P in range(0,6):
    Q = P/2+0.1
    x_tickposition.append(Q)

ax_2.bar(bar_position, bar_heights, 0.3)
ax_2.set_xticks(x_tickposition)
ax_2.set_xticklabels(x_tickslabels, rotation = 90)
ax_2.set_xlabel("Movies")
ax_2.set_ylabel("Total Views")
```

- The order of movies mostly watched by respondents is:

Star Wars V > Star Wars VI > Star Wars I > Star Wars IV > Star Wars II > Star Wars III

Except for the position interchange of Star Wars I and Star Wars IV, the sequence is the same. This means, viewers number is proportional to the quality of the movie

Our next step would be to divide the respondents on the basis of their gender and perform the similar analysis for each data to see if the results deviate

```python
male = star_wars[star_wars["Gender"] == "Male"]
rank = male.mean()
ax_3 = plt.subplot()
bar_heights = [rank["ranking_1"], rank["ranking_2"], rank["ranking_3"], rank["ranking_4"], rank["ranking_5"], rank["ranking_6"]]
bar_position =[]
x_tickslabels = ["Star Wars I", "Star Wars II","Star Wars III","Star Wars IV","Star Wars V","Star Wars VI"]
x_tickposition = []
for P in range(0,6):
    Q = P/2+0.1
    bar_position.append(Q)
    
for P in range(0,6):
    Q = P/2+ 0.1
    x_tickposition.append(Q)

ax_3.bar(bar_position, bar_heights, 0.3)
ax_3.set_xticks(x_tickposition)
ax_3.set_xticklabels(x_tickslabels, rotation = 90)
ax_3.set_xlabel("Movies")
ax_3.set_ylabel("Avg. Rating (Male)")
```

- Male Order: Star Wars V > Star Wars IV > Star Wars VI > Star Wars I > Star Wars II > Star Wars III  

- Female Order: Star Wars V > Star Wars VI > Star Wars I > Star Wars IV > Star Wars II > Star Wars III
